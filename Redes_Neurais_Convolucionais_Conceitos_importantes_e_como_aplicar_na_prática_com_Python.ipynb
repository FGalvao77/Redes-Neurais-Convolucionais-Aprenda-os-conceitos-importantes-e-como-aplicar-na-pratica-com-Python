## **Importando as bibliotecas**

```python
import numpy as np                # modelagem de dados
import matplotlib.pyplot as plt   # visualização gráfica
import matplotlib.image as mpimg  # visualização de imagem
import cv2                        # rede neural convolucional

import warnings                   # para ignorar alertas de erro
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# importando a imagem para o google colab
# utilizaremos uma biblioteca (library) disponibilizada pelo próprio google
from google.colab import files
uploaded = files.upload()
```

```python
# carregando a imagem na variável "image"
image = mpimg.imread('image1.png')

# visualizando a imagem
plt.imshow(image);
```

```python
# convertendo a imagem em "grayscale" (escala de cinza) para facilitar a identificações de padrões

# usaremos a função ".cvtColor" do "cv2", passando como argumentos 
# o objeto ("image") e a função que converter a imagem em escala de cinza ".COLOR_RGB2GRAY"
gray_image = cv2.cvtColor(image, 
                          cv2.COLOR_RGB2GRAY)

# visualizando a imagem convertida em escala de cinza
plt.imshow(gray_image, 
           cmap='gray');
```

## **Criando o kernel**

Um filtro muito comum para detectar bordas em imagens é o Operador Sobel.

- Pode ser usado para detectar padrões de intensidade nas imagens
- Aplicar esse filtro é como se tivéssimos tirando a derivada da imagem nas direções x e y.

***Filtro Sobel***

---


> https://pt.wikipedia.org/wiki/Filtro_Sobel

image.png

```python
# 3x3 array para detectar bordas

# instanciando  o "S_x" (array) - filtro vertical 
S_x = np.array([
                [-1, 0, 1], 
                [-2, 0, 2], 
                [-1, 0, 1]
])

# aplicando o filtro na imagem
imagem_filtrada = cv2.filter2D(gray_image, 
                               -1, 
                               S_x)

# visualizando a imagem depois de aplicado o filtro
plt.imshow(imagem_filtrada, 
           cmap='gray');
```

```python
# instanciando  o "S_y" (array) - filtro horizontal
S_y = np.array([
                    [-1, -2, -1], 
                    [0, 0, 0], 
                    [1, 2, 1]
])

# aplicando o filtro na imagem
imagem_filtrada2 = cv2.filter2D(gray_image, 
                                -1, S_y)

# visualizando a imagem depois de aplicado o filtro
plt.imshow(imagem_filtrada2, 
           cmap='gray');
```

```python
# instanciando  o "S_y" (array) - filtro horizontal
S_y = np.array([
                    [ 0, -1, 0], 
                    [ -1, 4, -1], 
                    [ 0, -1, 0]
])

# aplicando o filtro na imagem
imagem_filtrada3 = cv2.filter2D(gray_image, 
                                -1, 
                                S_y)

# visualizando a imagem depois de aplicado o filtro
plt.imshow(imagem_filtrada3, 
           cmap='gray');
```

```python
# instanciando  o "S_y" (array) - filtro horizontal
S_y = np.array([
                [ 0, -1, 0], 
                [ -1, 8, -1], 
                [ 0, -1, 0]
])

# aplicando o filtro na imagem
imagem_filtrada4 = cv2.filter2D(gray_image, 
                                -1, 
                                S_y)

# visualizando a imagem depois de aplicado o filtro
plt.imshow(imagem_filtrada4, 
           cmap='gray');
```

```python
# instanciando  o "S_y" (array) - filtro horizontal
S_y = np.array([
                [ 0, -1, 0], 
                [ -1, 4, -2], 
                [ 0, -1, 0]
])

# aplicando o filtro na imagem
imagem_filtrada5 = cv2.filter2D(gray_image, 
                                -1, 
                                S_y)

# visualizando a imagem depois de aplicado o filtro
plt.imshow(imagem_filtrada5, 
           cmap='gray');
```
